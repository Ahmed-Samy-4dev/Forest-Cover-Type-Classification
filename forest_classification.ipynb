```python
#ðŸ“¦ Step 1: Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

#ðŸ“‚ Step 2: Reading the dataset
columns = pd.MultiIndex.from_tuples([("","Elevation(m)"),("","Aspect"),("","Slope(deg)"),
                                     ("","Horizontal_Distance_To_Water(m)"),("","Vertical_Distance_To_Water(m)"),
                                     ("","Horizontal_Distance_To_Roadway(m)"),("","Hillshade_9am(0-255)"),
                                     ("","Hillshade_Noon(0-255)"),("","Hillshade_3pm(0-255)"),
                                     ("","Fire_Point_Horizontal_Distance(m)"),("Wilderness_Area","A"),
                                     ("Wilderness_Area","B"),("Wilderness_Area","C"),
                                     ("Wilderness_Area","D"),("Soil_Type","1"),("Soil_Type","2"),
                                     ("Soil_Type","3"),("Soil_Type","4"),("Soil_Type","5"),
                                     ("Soil_Type","6"),("Soil_Type","7"),("Soil_Type","8"),
                                     ("Soil_Type","9"),("Soil_Type","10"),("Soil_Type","11"),
                                     ("Soil_Type","12"),("Soil_Type","13"),("Soil_Type","14"),
                                     ("Soil_Type","15"),("Soil_Type","16"),("Soil_Type","17"),
                                     ("Soil_Type","18"),("Soil_Type","19"),("Soil_Type","20"),
                                     ("Soil_Type","21"),("Soil_Type","22"),("Soil_Type","23"),
                                     ("Soil_Type","24"),("Soil_Type","25"),("Soil_Type","26"),
                                     ("Soil_Type","27"),("Soil_Type","28"),("Soil_Type","29"),
                                     ("Soil_Type","30"),("Soil_Type","31"),("Soil_Type","32"),
                                     ("Soil_Type","33"),("Soil_Type","34"),("Soil_Type","35"),
                                     ("Soil_Type","36"),("Soil_Type","37"),("Soil_Type","38"),
                                     ("Soil_Type","39"),("Soil_Type","40"),("","Cover_Type(1-7)")])

df = pd.read_csv("./covtype.zip", compression="zip", header= None, names = columns)

#ðŸ§¹ Step 3: Explore & Clean the dataset
print("\nðŸ“Œ Exploring Dataset:")
display(df)

print("\nðŸ“Œ Exploring DataTypes of Features:\n")
print(df.info(),'\n')

print("\nðŸ“Œ Exploring Unique & Frequent Values in Features:")
display(df.describe(include='all'))

print("\nðŸ“Œ Null count:")
display(df.isnull().sum())
print("Total nulls:", df.isnull().sum().sum())

#ðŸ“Š Step 4: Visualize relationships
 # evaluating label imbalance in target
print("\nðŸ“Œ Visualizing Label Imbalance of Target(Cover Type):")
plt.figure(figsize=(6,4))
sns.histplot(df['']["Cover_Type(1-7)"], kde=True, bins=20)
plt.title("Cover Type Count")
plt.show()

print(f"\nConclusion:\n We'll need to check for bias in splits: since a dummy model would always tend to predict Cover Type '2' and get more than 50% accuracy")

 # evaluating wilderness count
print("\nðŸ“Œ Visualizing Wilderness Areas:")
plt.figure(figsize=(6,4))
plt.bar(df["Wilderness_Area"].columns,df["Wilderness_Area"].sum())
plt.title("Wilderness Area Count")
plt.show()

# evaluating soil type count
print("\nðŸ“Œ Visualizing Soil Types:")
plt.figure(figsize=(12,6))
plt.bar(df["Soil_Type"].columns,df["Soil_Type"].sum())
plt.title("Soil Type Count")
plt.show()

#ðŸŽ¯ Step 5: Assign Target & Features
y = df['']['Cover_Type(1-7)']
x = df.drop(columns=[("","Cover_Type(1-7)")])

x = x.sort_index(axis=1)
continuous = x.drop(columns=["Soil_Type","Wilderness_Area"])
one_hot = x[["Soil_Type","Wilderness_Area"]]

#ðŸ¤– Step 6: Training Random Forest Model on Continuous Features only
 # splitting continuous features into training & testing sets
x_train, x_test, y_train, y_test = train_test_split(continuous, y, test_size=0.25, random_state=1)

 # train a RandomForestClassifier model
rf = RandomForestClassifier(random_state=1, n_jobs=-1)
rf.fit(x_train, y_train.values.ravel())
importances = rf.feature_importances_

#ðŸ“Š Step 7: Visualizing the Importance of Continuous Features in prediction
print("\nðŸ“Œ Visualizing the Importance of Continuous Features:")
plt.figure(figsize=(8,4))
columns = x_train[''].columns
sns.barplot(x=importances, y= columns)
plt.title("Continuous Feature Importances (Random Forest)")
plt.show()

#ðŸ” Step 8: Predict & Visualize Outputs of Random Forest on continuous features only
 # predicting y values
prediction = rf.predict(x_test)

print("\nðŸ“Œ Random Forest Outputs using Continuous Features only:\n")
 # evaluating accuracy score
acc = accuracy_score(y_test.values, prediction)
print("Accuracy: %.3f" % acc)
 # store accuracy score
scores = [acc]

 # visualizing confusion matrix via heatmap
cm = confusion_matrix(y_test.values, prediction)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()

 # evaluating precision, recall, and f1 score via classification report
print("\nClassification Report:\n")
print(classification_report(y_test.values, prediction, zero_division=0))

print("\nConclusion:\n The model is consistent and is not driven by bias or memorization")
      
#ðŸ¤– Step 9: Training Random Forest Model on one_hot features only
 # splitting one_hot features into training & testing sets
x_train, x_test, y_train, y_test = train_test_split(one_hot, y, test_size=0.25, random_state=1)

 # train a RandomForestClassifier model
rf.fit(x_train, y_train.values.ravel())
importances = rf.feature_importances_

#ðŸ“Š Step 10: Identifying the Importance of one_hot in prediction
print("\nðŸ“Œ Visualizing the Importance of one_hot Features:")
plt.figure(figsize=(12,6))
columns = one_hot['Soil_Type'].columns.append(one_hot['Wilderness_Area'].columns)
sns.barplot(y=importances, x= columns)
plt.title("One_hot Feature Importances (Random Forest)")
plt.show()

#ðŸ” Step 11: Predict & Visualize Outputs of Random Forest on one_hot features only
prediction = rf.predict(x_test)

print("\nðŸ“Œ Random Forest Outputs using one_hot Features only:\n")
 # evaluating accuracy score
acc = accuracy_score(y_test.values, prediction)
print("Accuracy: %.3f" % acc)
 # store accuracy score
scores.append(acc)

 # visualizing confusion matrix via heatmap
cm = confusion_matrix(y_test.values, prediction)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()

 # evaluating precision, recall, and f1 score via classification report
print("\nClassification Report:\n")
print(classification_report(y_test.values, prediction, zero_division=0))

print("\nConclusion:\n The model is slightly biased by memorization: lower f1-score")

#ðŸ¤– Step 12: Training Random Forest Model on all features
 # splitting all features into training & testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)

 # train a RandomForestClassifier model
rf.fit(x_train, y_train.values.ravel())
importances = rf.feature_importances_

#ðŸ“Š Step 13: Identifying the Importance of all Features in prediction
print("\nðŸ“Œ Visualizing the Importance of all Features together (continuous & one_hot):")
plt.figure(figsize=(6,20))
columns = x_train[""].columns
columns = columns.append(one_hot['Soil_Type'].columns.append(one_hot['Wilderness_Area'].columns))
sns.barplot(x=importances, y= columns)
plt.title("Feature Importances (Random Forest)")
plt.show()

print("\nConclusion:\n One-hot features can increase performance: As Elevation's importance decreased when adding them(they took some of its importance)")

#ðŸ” Step 14: Predict & Visualize Outputs of Random Forest on all features
prediction = rf.predict(x_test)

print("\nðŸ“Œ Random Forest Outputs using all Features:\n")
 # evaluating accuracy score
acc = accuracy_score(y_test.values, prediction)
print("Accuracy: %.3f" % acc)
 # store accuracy score
scores.append(acc)

 # visualizing confusion matrix via heatmap
cm = confusion_matrix(y_test.values, prediction)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()

 # evaluating precision, recall, and f1 score via classification report
print("\nClassification Report:\n")
print(classification_report(y_test.values, prediction, zero_division=0))

print("\nConclusion:\n The model is slightly more precise and consistent")

#ðŸ¤– Step 15: Training XGBoost Model on all features
xgb = XGBClassifier(
    n_estimators=500,
    learning_rate=0.08,
    max_depth = 18,
    subsample=0.75,
    colsample_bytree = 0.75,
    random_state=1,
    eval_metric='mlogloss',
    tree_method="hist",
    device="cuda"
)

 # adjusting values to start with 0 to support XGBoost
y_train_fixed = y_train.values - 1
xgb.fit(x_train, y_train_fixed)

#ðŸ” Step 16: Predict & Visualize Outputs of XGBoost on all features
prediction = xgb.predict(x_test)

 # adding the subtracted value for comparing real values with predictions
prediction += 1

print("\nðŸ“Œ XGBoost using all features (with hypertunning):\n")
 # evaluating accuracy score
acc = accuracy_score(y_test.values, prediction)
print("Accuracy: %.3f" % acc)
 # store accuracy score
scores.append(acc)

 # visualizing confusion matrix via heatmap
cm = confusion_matrix(y_test.values, prediction)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()

 # evaluating precision, recall, and f1 score via classification report
print("\nClassification Report:\n")
print(classification_report(y_test.values, prediction, zero_division=0))


 # evaluating consistency of the accuracy of the model
print("\nðŸ“Œ XGBoost Cross Validation scores:\n")
cross_vals = cross_val_score(xgb, x_train, y_train_fixed, cv=5, scoring="accuracy")
print("Cross-val mean:", cross_vals.mean())
print("Fold scores:", cross_vals)

print("\nConclusion:\n The model is very high in accuracy & consistency")

#ðŸ“Š Step 17: Comparing Results
names = ['RandomForest(continuous)', 'RandomForest(one_hot)', 'RandomForest(all)', 'XGBoost(all)']

print("\nðŸ“Œ Comparing Accuracy of All Models:\n")

 # visualizing accuracy comparizon
plt.figure(figsize=(8,6))
plt.bar(names, scores, color=["#4C72B0", "#55A868", "#714EC4", "#C44E52"])
plt.title("Comparing Accuracy Scores across Models")
plt.xlabel('Model')
plt.ylabel("Accuracy Score")
plt.xticks(rotation=10)
plt.show()

print("\nðŸ“Œ Conclusion:\n - XGBoost with hypertunned parameters provided the highest accuracy while being very consistent" \
"\n - Continuous Features are the major predictors and thus are the most valuable data to capture precisely" \
"\n - RandomForest is better in prototyping and validating feature importance")
```
